1 Menor de dois pares: Escreva uma função que retorne o menor de dois números dados se ambos os números forem pares, mas retorna o maior se um dos dois for ímpar. Exemplo: menor_de_dois_pares(2,4) --> 2 menor_de_dois_inpares (2,5) --> 5

```python
def numeros(num1,num2):
    if num1 % 2 == 0 and num2 % 2 == 0:
        if num1 < num2:
            return num1
        else:
            return num2
    else:
        if num1 > num2:
            return num1
        else:
            return num2
            
print(numeros(2,4))
print(numeros(4,3))
```

2 Mesma letra: Escreva uma função que receba uma string com duas palavras e retorne True se ambas palavras começarem com a mesma letra. Exemplo: mesma_letra('Cão covarde') -> True mesma_letra('Vira Lata') -> False

```python
def primeira_letra(arg):
    palavras = arg.title().split()
    if palavras[0][0] == palavras[1][0]:
        return True
    else:
        return False
    
print (primeira_letra('Matheus Sacho'))
print (primeira_letra('Cão covarde'))
```

3 Mestre Yoda: Dada uma sentença, a função deve retornar a sentença com as palavras na ordem reversa. Exemplo: mestre_yoda('Eu estou em casa') --> 'casa em estou Eu' mestre_yoda('Estamos prontos') --> 'prontos Estamos'

```python
string = input('Insira uma frase:')

def mestre_yoda(string):
    palavras = string.split()
    palavras.reverse()
    yoda = ""
    for i in range(0, len(palavras) - 1):
        palavras[i] = palavras[i] + " "
    for palavra in palavras:
        yoda += palavra
    return yoda

print(f"mestre_yoda('{string}') --> '{mestre_yoda(string)}'")
```

4 Tem 33: Faça uma função que retorne True se, dada uma lista de inteiros, houver em alguma posição da lista um 3 do lado de outro 3. Exemplo: tem_33([1,3,3]) --> True tem_33([1,3,1,3]) --> False tem_33([3,1,3]) --> False

```python
def tem_33(arg):
    temTres = False
    for i in arg:
        if i == 3:
            if temTres is True:
                return True
            else:
                temTres = True
        else:
            temTres = False
    else:
        return False
    
print (tem_33([1,3,3]))
print (tem_33([1,3,1,3]))
print (tem_33([3,1,3]))
```

5 Blackjack: Faça uma função que receba 3 inteiros entre 1 e 11. Se a soma deles for menor que 21, retorne o valor da soma. Se for maior do que 21 e houver um 11, subtraia 10 da soma antes de apresentar o resultado. Se o valor da soma passar de 21, retorne ‘ESTOUROU’. Exemplo:

blackjack(5,6,7) --> 18 blackjack(9,9,9) --> 'ESTOUROU' blackjack(9,9,11) --> 19

```python
def soma(num1,num2,num3):
    x= num1 + num2 + num3
    
    if x <= 21:
        return x
    
    else:
        if num1 == 11 or num2 == 11 or num3 == 11:
            x -= 10
        if x > 21:
            return 'Estourou'
        else:
            return x
        
print (soma(1,8,7))
print (soma(10,8,10))
print (soma(11,11,8))
```
